```python
import pandas as pd
from pandas import read_csv
#import os 
#from glob import glob

data='/Users/PC/Desktop/pnad_educacao_2019/dados_pnad_2019.csv'
pnad_2019_edu = read_csv (data, sep=',', encoding='latin1')
pnad_2019_edu
```

```python
pnad_2019_edu.info()
```

```python
#Analfabetos Funcionais
# 7 e 8 representam as pessoas ficaram na escola atÃ© o priemiro grau e o 1Âº do supletivo
analf_funci = [7,8]
analf_funci= pnad_2019_edu[pnad_2019_edu['V3009A'].isin(analf_funci)].copy()
analf_funci
```

```python
# Nomeando as regiÃµes:
regi = ['Norte','Nordeste','Centro-Oeste','Sudeste','Sul']
df_regi = pd.DataFrame(regi, columns=['Regiao'])

# os cÃ³digos UF sÃ£o: 11-17 Norte, 21-29 Nordeste, 31-35 Sudeste, 41-43 Sul, 50-53 Centro-Oeste
uf_regi = {
    11: 'Norte', 12: 'Norte', 13: 'Norte', 14: 'Norte', 15: 'Norte', 16: 'Norte', 17: 'Norte',
    21: 'Nordeste', 22: 'Nordeste', 23: 'Nordeste', 24: 'Nordeste', 25: 'Nordeste', 26: 'Nordeste', 27: 'Nordeste', 28: 'Nordeste', 29: 'Nordeste',
    31: 'Sudeste', 32: 'Sudeste', 33: 'Sudeste', 34: 'Sudeste', 35: 'Sudeste',
    41: 'Sul', 42: 'Sul', 43: 'Sul',
    50: 'Centro-Oeste', 51: 'Centro-Oeste', 52: 'Centro-Oeste', 53: 'Centro-Oeste'
}
uf_regi
```

```python
# aqui uno a tabela uf a regiÃ£o e contabilizo a quntidade de analfabetos funcionais em cada regiÃ£o
analf_funci.loc[:, 'Regiao'] = analf_funci['UF'].map(uf_regi)
df_total = analf_funci.groupby('Regiao').size().reset_index(name='Quanti')
df_total = df_regi.merge(df_total, on='Regiao', how='left').fillna(0)
df_total['Quanti'] = df_total['Quanti'].astype(int)

df_total
```

```python
#plotanto um grÃ¡fico que representa a quantidade de analfabetos por regiÃ£o
import matplotlib.pyplot as plt

df_total.plot(kind='bar', x='Regiao', y='Quanti', legend=False)
plt.title('Analfabetos Funcionais por RegiÃ£o')
plt.xlabel('Regiao')
plt.ylabel('Quantidade')
plt.show()
```

```python
# analfabetos funcionais por sexo 
analf_funci_sexo = analf_funci['V2007'].value_counts().rename(index={1: 'Masculino', 2: 'Feminino'})
analf_funci_sexo
```

```python
# plotando o grafico que vai representar a quantidade de analfabetos homens e mulheres:
analf_funci_sexo.plot(kind='bar', color=['blue', 'pink'])
plt.title('DistribuiÃ§Ã£o de Analfabetos Funcionais por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)  
plt.show()
```

```python
#Analfabetos funcionais por cor e raÃ§a
analf_funci_raca = analf_funci['V2010'].value_counts().rename(
    index={
    1: 'Branca',
    2: 'Preta',
    3: 'Amarela',
    4: 'Parda',
    5: 'IndÃ­gena',
    9: 'Ignorado'
    })

analf_funci_raca
```

```python
#plotando o grafico
analf_funci_raca.plot(kind='bar', color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6'])
plt.title('DistribuiÃ§Ã£o por Cor ou RaÃ§a')
plt.xlabel('Cor/RaÃ§a')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)  
plt.show()
```

```python
# Filtrando faixa por IDADE
bins = [0, 18, 30, 40, 50, 60, 70, 120]  # Ajustando para incluir o limite superior
labels = ['0-17', '18-29', '30-39', '40-49', '50-59', '60-69', '70+']
analf_funci.loc[:, 'Faixa_Etaria'] = pd.cut(analf_funci['V2009'], bins=bins, labels=labels, right=False)
analf_funci_faixa_etaria = analf_funci['Faixa_Etaria'].value_counts(sort=False)
print(analf_funci_faixa_etaria)
print(f'Total de pessoas da base de dados: {analf_funci.shape[0]}')  
```
